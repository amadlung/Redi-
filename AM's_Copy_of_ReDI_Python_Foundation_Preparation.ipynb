**V. Functions**

To Do: Create a function that takes 1 string as an argument and adds a exclamation point to it, and returns the result.

```python

#define function called "Exclaim" that takes a string and then prints the string and adds an exclamation mark

def exclaim(what="string"):
    print(what,"!")

#call function with default
exclaim()
#call function with any string, here as example "Take this"
exclaim(what="Take this")

```

```python
# same function but removing the space between the argument and the !
def exclaim_1(what="string"):
    print(what + "!")
exclaim_1 ()
```

```python
# same function but removing the space between the argument and the ! and returning result
def exclaim_2 (what):
    return (what + "!")

result = exclaim_2 ("and this")
print (result)

```
